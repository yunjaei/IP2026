```python

import numpy as np
import cv2
```

```python
def nothing(x):
    pass
```

```python
drawing = False # true if mouse is pressed
ix,iy = -1,-1
cx, cy = -1, -1
org_img = cv2.imread('sea.jpg')
img = org_img.copy()
# mouse callback function
def draw_circle(event,x,y,flags,param):
    global ix,iy,cx,cy, drawing,mode, img
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix,iy = x,y
    elif event == cv2.EVENT_MOUSEMOVE:
        cx, cy = x, y                  
    elif event == cv2.EVENT_LBUTTONUP:
        ix,iy = x,y
        cx, cy = x, y 
        drawing = False
```

```python
cv2.namedWindow('image')
cv2.createTrackbar('value','image',0,255,nothing)
cv2.setMouseCallback('image', draw_circle)
font = cv2.FONT_HERSHEY_SIMPLEX
org_img[83:176, 123:298, 2] = 255
while(1):
    img = org_img.copy()
    value = cv2.getTrackbarPos('value','image')
    if drawing == True:
        cv2.rectangle(img,(ix,iy),(cx,cy),(0,255,0),-1)
    cv2.putText(img,f"Mouse position : ({ix},{iy})-({cx},{cy})-{value}",(10,30), font, 0.7,(255,255,255),2,cv2.LINE_AA)
    cv2.imshow('image',img)
    k = cv2.waitKey(1) & 0xFF
    if  k == 27:
        break
cv2.destroyAllWindows()
```

```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
